```python
import numpy as np
import pandas as pd

# Getting the data
pop_census= pd.read_csv("../datasets/US_Census_Population_2022.csv")
pop_census = pop_census.iloc[2:]
pop_census.head()
```

```python
def clear_geoid_metadata(census_data):
    return census_data.GEO_ID.map(lambda id : id[9:])
```

```python
# Getting rid of metadata in front of the 11-digit GEOID
parsed_geoids = clear_geoid_metadata(pop_census)
print(parsed_geoids)
```

```python
# Update Dataframe
pop_census.GEO_ID = parsed_geoids
pop_census.head()
```

```python
# Relevant Population Columns : TOTAL POPULATION
POPULATION_COLUMNS = ["DP05_0001E"]
id_and_pop = pop_census[["GEO_ID"] + POPULATION_COLUMNS]
id_and_pop.head()
```

```python
# getting the poverty data
pov_census = pd.read_csv("../datasets/US_Census_Poverty_2022.csv")
pov_census = pov_census.iloc[2:]
pov_census.GEO_ID = clear_geoid_metadata(pov_census)
pov_census.head()
```

```python
POVERTY_COLUMNS = ["S1701_C01_001E"]
pov_frame = pd.DataFrame(pov_census[POVERTY_COLUMNS], columns=POVERTY_COLUMNS)
coa_dataset = pd.concat([id_and_pop, pov_frame], axis=1)
coa_dataset.head()
```

```python
# Getting our zipcode & wastewater data
wastewater_data = pd.read_csv("../datasets/UGS_Houston_Wastewater_Data.csv")
wastewater_data.head()
```

```python
import http.client
```

```python
REQUEST_PATHS = [BASE_URL_PATH + "?" + ZIP_TO_TRACT_TYPE + "&query=", BASE_URL_PATH + "?" + TRACT_TO_ZIP_TYPE + "&query="]
REQUEST_TYPES = ["ztot", "ttoz"] # "zip to tract" (0), and "tract to zip" (1)

# payload = {}
HEADER = {'Authorization' : "Bearer " + USPS_TOKEN}

# returns a Zip to tract request path given the current zipcode
def get_query_string(input_zipcode, type):
    if type is None: 
        print("get_query_string(); Type cannot be None. Returning None.")
        return None

    index = 0
    while (index < len(REQUEST_TYPES) and type != REQUEST_TYPES[index]):
        index += 1

    return REQUEST_PATHS[index] + str(input_zipcode)

def attempt_request(HTTPS_CONN, qs, total_attempts = 1):
    attempts = 0
    success = False
    payload = {}
    while attempts < total_attempts and not success:
        try:
            HTTPS_CONN.request("GET", qs, payload, HEADER)
            response = HTTPS_CONN.getresponse()
            response_decode = response.read().decode("utf-8")
            response_json = json.loads(response_decode)
            json_results = response_json["data"]["results"]
            return json_results
        except (json.decoder.JSONDecodeError):
            print(f"JSONDecodeError {attempts + 1}")
            attempts += 1
    return None

# Extract the JSON file associated with the current zipcode from the USPS crosswalk API
def get_zip_json(HTTPS_CONN, zip):
    payload = {}

    # Making the request and extracting the json
    qs = get_query_string(zip, type="ztot")
    print(qs)
    return attempt_request(HTTPS_CONN, qs, 5)

# Extract data from the relevant columns from the geoids that map with the zipcode
def make_zip_row(zip_json, geoid_df, columns_to_extract):
    buffer = np.array([0] * len(columns_to_extract))
    print(f"Number of tracts: {len(zip_json)}")
    for tract in zip_json:
        GEO_ID = tract["geoid"]
        RES_RATIO = tract["res_ratio"]

        geoid_row = geoid_df.loc[geoid_df["GEO_ID"] == GEO_ID][columns_to_extract]
        try:
            geoid_row = geoid_row.values[0].astype(float)
            buffer = np.add(buffer, np.multiply(RES_RATIO, geoid_row))
        except (IndexError):
            print("IndexError? has occured")
            print(f"GEOID: {GEO_ID}")
            print(geoid_row)
    return buffer

# extract the specified columns in columns_to_extract from a given dataframe with a GEOID column
def make_zip_columns(zip_series, geoid_df, columns_to_extract):
    if geoid_df is None: 
        print("make_zip_columns(); function arguments cannot be None. Returning None.")
        return None
    
    CONN = http.client.HTTPSConnection(BASE_URL_DOMAIN)
    out_columns = [] * len(zip_series)
    
    for count, zip in enumerate(zip_series):
        print(count)
        zip_json = get_zip_json(CONN, zip)
        out = make_zip_row(zip_json, geoid_df, columns_to_extract)
        out_columns.append(out)

    # Make the request and retrieve the json file
    # Iterate through the 'results' section and get each geoid
    # Use our geoid_dif and locate that geoid's row
    # extract the necessary information from that row and partition it based on the residential ratio
    
    CONN.close()
    return out_columns

GEO_ID_COLUMNS_TO_EXTRACT = coa_dataset.columns[1:]
new_zip_columns = make_zip_columns(ww_data["Zipcode"], coa_dataset, GEO_ID_COLUMNS_TO_EXTRACT)
new_zip_cols_df = pd.DataFrame(new_zip_columns, columns=GEO_ID_COLUMNS_TO_EXTRACT)
zip_complete = pd.concat([ww_data, new_zip_cols_df], axis=1)
```
